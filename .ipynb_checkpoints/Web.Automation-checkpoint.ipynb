Search prices on Google Shopping and Buscape

```python
#start setup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
import time
import win32com.client as win32

nav = webdriver.Chrome()

#import search terms
search_terms = pd.read_excel("search-terms.xlsx")
display(search_terms)
```

Define Search Function for Google Shopping e Buscape

```python
def search_google_shopping(nav, product, banned_terms, min_price, max_price):

    #open google
    nav.get("https://www.google.com")

    #process search terms
    product = product.lower()
    banned_terms = banned_terms.lower()
    banned_terms_list = banned_terms.split(" ")
    product_list = product.split(" ")
    min_price = float(preco_minimo)
    max_price = float(preco_maximo)
    
    #search product term
    nav.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(product)
    nav.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)

    #click on Shopping bar
    elements = nav.find_elements(By.CLASS_NAME, 'hdtb-mitem')
    for item in elements:
        if "Shopping" in item.text:
            item.click()
            break
            
    #create a list for the search results
    results_list = nav.find_elements(By.CLASS_NAME, 'sh-dgr__content')
    
    #for each item in result_list, checks if it match with the search_terms list
    
    offer_list = []

    for result in results_list:
        product = result.find_element(By.CLASS_NAME, 'Xjkr3b').text
        product = product.lower()

        #checks for banned terms
        is_a_banned_term = False
        for term in banned_terms_list:
            if term in product:
                is_a_banned_term = True
                
        #checks the product terms
        have_all_product_terms = True
        for term in product_list:
            if term not in product:
                have_all_product_terms = False
                
        
        if not is_a_banned_term and have_all_product_terms:
            try:
                price = result.find_element(By.CLASS_NAME, 'a8Pemb').text
                price = price.replace("R$", "").replace(" ", "").replace(".", "").replace(",", ".")
                price = float(price)
            except:
                    continue

            if min_price <= price <= max_price: #checks if the price range corresponds to the established values       
                link_element = result.find_element(By.CLASS_NAME, 'aULzUe')
                parent_element= link_element.find_element(By.XPATH, '..')
                link = parent_element.get_attribute('href')
                offer_list.append((product, price, link))
    return offer_list
        
def search_buscape(nav, product, banned_terms, min_price, max_price):
    
    #open buscape
    nav.get("https://www.buscape.com.br/")
    
    #process search terms
    product = product.lower()
    banned_terms = banned_terms.lower()
    banned_terms_list = banned_terms.split(" ")
    product_list = product.split(" ")
    min_price = float(preco_minimo)
    max_price = float(preco_maximo)
    
    #search term
    nav.find_element(By.CLASS_NAME, 'search-bar__text-box').send_keys(product, Keys.ENTER)
    
    time.sleep(5)  
    results_list = nav.find_elements(By.CLASS_NAME, 'Cell_Content__1630r')
    
    offer_list = []    
    for result in results_list:
        try:
            price = result.find_element(By.CLASS_NAME, 'CellPrice_MainValue__3s0iP').text
            product = resultado.get_attribute('title')
            product = product.lower()
            link = result.get_attribute('href')
            
            #checks for banned terms
            is_a_banned_term = False
            for term in banned_terms_list:
                if term in product:
                    is_a_banned_term = True

            #checks the product terms
            have_all_product_terms = True
            for term in product_list:
                if term not in product:
                    have_all_product_terms = False


            if not is_a_banned_term and have_all_product_terms:
                try:
                    price = price.replace("R$", "").replace(" ", "").replace(".", "").replace(",", ".")
                    price = float(price)
                except:
                    continue

                if min_price <= price <= max_price: #verifica se o preço está de acordo           
                    offer_list.append((product, price, link))
                        
        except:
            pass
        
    return offer_list


def send_mail(towho, attach):
    
    outlook = win32.Dispatch('outlook.application')
    mail = outlook.CreateItem(0)
    mail.To = towho
    mail.Subject = 'Products found in your price range'
    mail.HTMLBody = f"""
    <p>To whom it may concern,</p>
    <p>We found somes itens in your price range. For details, follow the table below:</p>
    {attach.to_html(index=False)}
    <p>Please let me know if you have any questions.</p>
    <p>Best regards.</p>
    <p>Mr. Robot</p>
    """
    
    mail.Send()
```

Variable construction

```python
offer_table = pd.DataFrame()

for line in search_terms.index:
    product = search_terms.loc[line, "Name"]
    banned_terms = search_terms.loc[line, "Banned Terms"]
    min_price = search_terms.loc[line, "Min. Price"]
    max_price = search_terms.loc[line, "Max. Price"]
    
    google_shopping_offer_list = search_google_shopping(nav, product, banned_terms, min_price, max_price)
    if google_shopping_offer_list:
        google_shopping_table = pd.DataFrame(google_shopping_offer_list, columns=['Product', 'Price', 'Link'])
        offer_table = offer_table.append(google_shopping_table)
    else:
        google_shopping_table = None

    buscape_offer_list = search_buscape(nav, product, banned_terms, min_price, max_price)
    if buscape_offer_list:
        buscape_table = pd.DataFrame(buscape_offer_list, columns=['Product', 'Price', 'Link'])
        offer_table = offer_table.append(buscape_table)
    else:
        buscape_table = None
    
```

Exporting results to Excel file

```python
offer_table = offer_table.reset_index(drop=True)
offer_table.to_excel("Offer_list.xlsx", index=False)
```
